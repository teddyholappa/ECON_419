```python
req_list = [jan_2020, feb_2020, mar_2020, apr_2020, may_2020, jun_2020, jul_2020, aug_2020, sep_2020, oct_2020,
            nov_2020, dec_2020, jan_2021, feb_2021, mar_2021]
names = ['jan_2020', 'feb_2020', 'mar_2020', 'apr_2020', 'may_2020', 'jun_2020', 'jul_2020', 'aug_2020', 'sep_2020', 'oct_2020',
            'nov_2020', 'dec_2020', 'jan_2021', 'feb_2021', 'mar_2021']
df_list = []
```

```python
states = ['AL','AK','AZ','AR','CA','CO','CT','DE',
         'DC','FL','GA','HI','ID','IL','IN','IA','KA',
         'KY','LA','ME','MD','MA','MI','MN','MS','MO',
         'MT','NE','NV','NH','NJ','NM','NY','NC','ND',
         'OH','OK','OR','PA','RI','SC','SD','TN','TX','UT',
         'VT','VA','WA','WV','WI','WY']
states_dict = {
        '1':1,'2':2,'4':3,'5':4,'6':5,'8':6,'9':7,
        '10':8,'11':9,'12':10,'13':11,'15':12,'16':13,'17':14,
        '18':15,'19':16,'20':17,'21':18,'22':19,'23':20,'24':21,
        '25':22,'26':23,'27':24,'28':25,'29':26,'30':27,'31':28,
        '32':29,'33':30,'34':31,'35':32,'36':33,'37':34,'38':35,
        '39':36,'40':37,'41':38,'42':39,'44':40,'45':41,'46':42,
        '47':43,'48':44,'49':45,'50':46,'51':47,'53':48, '54':49,
        '55':50,'56':51
}
m_status = ['Married Spouse Present', 'Married Spouse Absent',
            'Widowed','Divorced','Separated','Never Married']
education = ['No Diploma','High school','Some college','Associates','Bachelors','Masters','PhD']
race = ['White','Black','HP','Asian','American Indian']
citizenship = ['Native Citizen','Naturalized Citizen','Non-Citizen']
```

```python
from sklearn.preprocessing import LabelBinarizer
def one_hot_encode(df,columns, col_name):
    y = LabelBinarizer().fit_transform(df[col_name])
    one_hot = pd.DataFrame(index=df.index,data=y,columns=columns)
    return one_hot
```

```python
def drop_out_of_universe(df):
    df = df[df.PESEX != '-1']
    df = df[df.PEMARITL != '-1']
    #Removing pre-high school values
    '''
    1 - no high school diploma
    2-high school grad
    3 - Some college
    4 - Associates degree
    5 - Bachelor's
    6 - Master's
    7 - PHD or professional school
    '''
    bad_educ_vals = ['31','32','33','34','35','36','37'] #replace these with -1
    df['PEEDUCA'] = df['PEEDUCA'].replace(bad_educ_vals,'-1')
    df['PEEDUCA'] = df['PEEDUCA'].replace(['38','39','40','41','42','43','44','45','46'],[1,2,3,4,4,5,6,7,7])
    df = df[df.PEEDUCA != '-1']
    df = df[df.PEEDUCA != -1]
    bad_race_values = ['26','25','24','23','22','21','20','19','18','17',
                      '16','10','5','13','14','15']
    df['PTDTRACE'] = df['PTDTRACE'].replace(bad_race_values,-1)
    '''
    White = 1
    Black = 2
    HP = 3
    Asian = 4
    AI = 5
    '''
    df['PTDTRACE'] = df['PTDTRACE'].replace(['1','2','3','4','6','7','8','9','10','11','12'],[1,2,5,4,2,5,4,3,2,2,3])
    df = df[df.PTDTRACE != '-1']
    df = df[df.PTDTRACE != -1]
    df['PRCITSHP'] = df['PRCITSHP'].replace(['1','2','3','4','5'],[1,1,1,2,3])
    df = df[df.PRCITSHP != '-1']
    df = df[df.PREXPLF != '-1']
    df['STATE'] = df['STATE'].replace(states_dict)
    df = df.drop(columns=['PWSSWGT','PWMCMPWGT'])
    return df
```

```python
cols = ['PWSSWGT','PWMCMPWGT','PESEX','PEMARITL','PEEDUCA',
        'PTDTRACE','PRCITSHP','PREXPLF','STATE']
```

```python
df_list = []
for i in range(len(req_list)):
    #Begin the one_hot_encoding process
    request = req_list[i]
    response = requests.get(request)
    name = names[i] +".csv"
    query = json.loads(response.text)[1:]
    df = pd.DataFrame(columns=cols, data=query)
    #Drop out of universe values
    df = drop_out_of_universe(df)
    #PESEX
    pesex_one = pd.to_numeric(df['PESEX']) - 1
    #PEMARITL
    premaritl_one = one_hot_encode(df,m_status,'PEMARITL')
    #PEEDUCA - this one is more complicated
    #We don't necessarily need to one hot encode this, as it has a scale!!!
    #peeduca_one = one_hot_encode(df,education,'PEEDUCA')
    peeduca_one = df['PEEDUCA']
    #PTDTRACE
    ptdtrace_one = one_hot_encode(df, race, 'PTDTRACE')
    #PRCITSHP
    prcitshp_one = one_hot_encode(df,citizenship,'PRCITSHP')
    #PREEXPLF - already one hot encoded
    preexplf_one = pd.to_numeric(df['PREXPLF']) - 1
    #STATE
    states_one = one_hot_encode(df,states,'STATE')
    #Create the one hot encoded dataframe!
    result = pd.concat([preexplf_one,pesex_one,peeduca_one, premaritl_one,ptdtrace_one,
                       prcitshp_one,states_one],axis=1)
    #Add it to the df_list
    df_list.append(result)
    result.to_csv('all_data/' + name)
```

```python
class RegData():
    def _init_(self,name,reg):
        self.name = name
        self.reg = reg
    
```

```python
#Time to regress!
regression_list = []
import statsmodels.api as sm
for i in range(len(df_list)):
    df = df_list[i]
    df = df.drop(columns=['AK','Non-Citizen','HP','Separated'])
    #Value for how much training data to use - say, 90%
    idx = int(df.shape[0]*.9)
    X = df.drop('PREXPLF',axis=1)
    Y = df['PREXPLF']
    logistic_regression = sm.Logit(Y,X).fit_regularized(disp=0)
    regression_list.append(logistic_regression)
    
    
```

```python
reg = regression_list[1]
reg.summary()
```

```python
#Plot the state's coefficients for each time period
cal_coef = []
tx_coef = []
ct_coef = []
ny_coef = []
sd_coef = []
ne_coef = []
ia_coef = []
fl_coef = []
vt_coef = []
dc_coef = []
for reg in regression_list:
    coef = reg.params['CA']
    tx = reg.params['TX']
    ct = reg.params['CT']
    ny = reg.params['NY']
    sd = reg.params['SD']
    ne = reg.params['NE']
    ia = reg.params['IA']
    fl = reg.params['FL']
    vt = reg.params['VT']
    dc = reg.params['DC']
    cal_coef.append(coef)
    tx_coef.append(tx)
    ct_coef.append(ct)
    ny_coef.append(ny)
    sd_coef.append(sd)
    ne_coef.append(ne)
    ia_coef.append(ia)
    fl_coef.append(fl)
    vt_coef.append(vt)
    dc_coef.append(dc)
    
```

```python
#Create a DF of all the state's coefficents
columns = states
idx = names
states_df = pd.DataFrame(columns=columns, index=idx)
for state in states:
    coef_list = []
    for reg in regression_list:
        coef = reg.params[state]
        coef_list.append(coef)
    states_df[state] = coef_list

from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters = 10).fit(states_df.T)
kmeans = kmeans.predict(states_df.T)
#print(kmeans)
clustering = pd.DataFrame(data=kmeans,index=states,columns=['Label'])
clustering = clustering.sort_values(by=['Label'])
```
